```python
# keys() --> Returns all keys in the dictionary.

student = {"name": "Dipam", "age": 29}

print(student.keys())
```

```python
# values() --> Returns all values.

print(student.values())
```

```python
# items() --> Returns key-value pairs as tuples.

print(student.items())
```

```python
# get(key, default_value) --> Safe key access. 
# ðŸ‘‰ Use when you're not sure if the key exists. Prevents errors.

print(student.get("city", "Not Available"))
```

```python
# setdefault() 
# If key exists â†’ return value
# If key doesnâ€™t exist â†’ create it with default value
# Best when adding new keys safely without overwriting existing ones.

student.setdefault("city", "Bangalore")
print(student)
```

```python
# copy() --> Creates a shallow copy of dictionary (important for avoiding reference issues).

new_dict = student.copy()

# ðŸ‘‰ Use when you want same data but not linked to original.
```

```python
# clear() --> Removes ALL items.

student.clear()
print(student)
```

```python
#1ï¸âƒ£ Using in keyword (BEST & MOST COMMON)
student = {
    "name": "Dipam",
    "age": 29,
    "skills": ["Python", "CAE"]
}

if "age" in student:
    print("Age exists")
else:
    print("Age not found")
```

```python
#2ï¸âƒ£ Using not in
if "city" not in student:
    print("City not available")

#3ï¸âƒ£ Why NOT student["city"] directly?
print(student["city"]) #âŒ Error:
```

```python
#4ï¸âƒ£ Practical Example (Real Usage)
data = {"status": "success"}

if "error" in data:
    print("Error occurred")
else:
    print("No error, continue processing")

#5ï¸âƒ£ Interview Tip (IMPORTANT)

#ðŸ‘‰ Correct answer when asked:
#How do you check if a key exists in a dictionary?
#âœ… Answer:
#Using the in keyword.
```

```python
#1ï¸âƒ£ What is Dictionary Comprehension?

#A short and clean way to create dictionaries using a loop.

#Normal way:
squares = {}

for i in range(1, 6):
    squares[i] = i * i

print(squares)

#Dictionary comprehension way:
squares = {i: i*i for i in range(1, 6)}
print(squares)
```

```python
#2ï¸âƒ£ Syntax (Must Remember)
#{key: value for item in iterable}

#3ï¸âƒ£ Example: Convert List to Dictionary
names = ["Dipam", "Rahul", "Amit"]

lengths = {name: len(name) for name in names}
print(lengths)
```

```python
#4ï¸âƒ£ Dictionary Comprehension with Condition
numbers = range(1, 11)

even_squares = {n: n*n for n in numbers if n % 2 == 0}
print(even_squares)
```

```python
#5ï¸âƒ£ Modify Existing Dictionary
marks = {"math": 78, "science": 92, "english": 65}

updated = {subject: mark+5 for subject, mark in marks.items()}
print(updated)
```

```python
#6ï¸âƒ£ Real Data Science Example (Important)
scores = {"A": 90, "B": 45, "C": 82}

result = {k: "Pass" if v >= 50 else "Fail" for k, v in scores.items()}
print(result)
```
