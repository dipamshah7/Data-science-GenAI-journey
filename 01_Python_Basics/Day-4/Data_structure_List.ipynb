```python
# Finding all indexes of a value in a list
numbers = [2, 5, 2, 7, 9, 2]

indexes = [i for i, v in enumerate(numbers) if v == 2]

print(indexes)
```

```python
# List Methods: append, insert, extend, pop, remove, sort, reverse

items = [10, 20, 30]

items.append(40) # [10, 20, 30, 40]
items.insert(1, 15) # [10, 15, 20, 30, 40]
items.extend([50, 60]) # [10, 15, 20, 30, 40, 50, 60]

popped = items.pop() # removes and returns last item (60)
items.remove(30) # removes first occurrence of 30
items.sort() # sorts the list in ascending order
items.reverse() # reverses the list

print(items)
print("Popped:", popped)
```
