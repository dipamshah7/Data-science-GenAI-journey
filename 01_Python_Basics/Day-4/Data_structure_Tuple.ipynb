```python
# First topic: Tuples

coordinates = (10.5, 20.3, 30.7)

print(coordinates)
print(type(coordinates))

print(coordinates[0])
print(coordinates[-1])

print(len(coordinates))
```

```python
# Tuple unpacking
point = (5, 10)

x, y = point   # unpacking

print("x =", x)
print("y =", y)
```

```python
# Functions returning multiple values
def calc(a, b):
    return a+b, a*b, a/b

result = calc(10, 2)
print(result)

# Unpacking the returned tuple
add, mul, div = calc(10,2)
print(add, mul, div)
```

```python
# Tuple methods
t = (1, 2, 2, 3, 2)

print(t.count(2))
print(t.index(3))
```

```python
# Finding specific element indexes of a value in a tuple
t = (1, 2, 2, 3, 2)

indexes = []

for i, value in enumerate(t):
    if value == 2:
        indexes.append(i)

print(indexes)


numbers = [5, 10, 15]

for i, n in enumerate(numbers):
    print(i, n)
```
