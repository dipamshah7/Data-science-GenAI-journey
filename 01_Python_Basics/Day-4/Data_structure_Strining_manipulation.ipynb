```python
#1️⃣ Basic but Important String Operations
text = "  Data Science Learning  "

#Length
print(len(text))

#Lower/Upper Case
print(text.lower())
print(text.upper())

print(text.strip())     # both sides
print(text.lstrip())    # left only
print(text.rstrip())    # right only
```

```python
#2️⃣ Finding Characters / Words

sentence = "python is powerful"

print(sentence.count("p"))

print(sentence.find("power"))

print(sentence.index("python"))

```

```python
#3️⃣ Replace (Data Cleaning Core)

data = "salary$1000"
clean = data.replace("$", "")
print(clean)
```

```python
#4️⃣ Startswith / Endswith (Filtering Data)

filename = "report.csv"

print(filename.startswith("rep"))
print(filename.endswith(".csv"))
```

```python
#5️⃣ Split & Join (EXTREMELY IMPORTANT)
#split()

skills = "Python,ML,GenAI,SQL"
skill_list = skills.split(",")
print(skill_list)

#join()
joined = " | ".join(skill_list)
print(joined)
```

```python
#6️⃣ String Slicing (Revision + Power Use)

word = "ENGINEER"

print(word[0:4])   # ENGI
print(word[-3:])   # EER
print(word[::-1])  # Reverse
```
