```python
#2️⃣ Creating a Set
numbers = {1, 2, 3, 4, 5}
print(numbers)
print(type(numbers))

#3️⃣ Set Automatically Removes Duplicates
data = {1, 2, 2, 3, 4, 4, 5}
print(data)
```

```python
#4️⃣ Creating Set from List (Very Important)
marks = [70, 80, 70, 90, 80]
unique_marks = set(marks)

print(unique_marks)
```

```python
#5️⃣ Adding & Removing Elements
#Add
numbers.add(6)
print(numbers)

#Remove (Error if not present)
numbers.remove(3)

#Discard (Safe — no error)
numbers.discard(10)

#6️⃣ Checking Membership (Very Fast)
print(3 in numbers)
print(10 in numbers)
```

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)   # union
print(a & b)   # intersection
print(a - b)   # difference


#7️⃣ Set Comprehension (Like List/Dict Comprehension)
#Example: Even numbers from 1–10
even_set = {x for x in range(1, 11) if x % 2 == 0}
print(even_set)

#Example: Unique squared values
nums = [1, 2, 2, 3, 3, 4]
squares = {n*n for n in nums}
print(squares)
```

```python
names = ["Dipam", "Amit", "Dipam", "Ravi", "Amit"]
unique_names = set(names)
print(unique_names)
```
