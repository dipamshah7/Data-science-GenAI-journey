```python
# ‚úÖ Exercise 7 ‚Äî Find Maximum of Three (without using max())

def maximum_of_three(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(maximum_of_three(25, 10, 5))
```

```python
# ‚úÖ Exercise 10 ‚Äî Function Returning Multiple Values
# Task:
# Create a function student_report(name, maths, science, english) that:
# Must Do:
    # Calculate total marks
    # Calculate average
    # Return all three values ‚Üí name, total, average
# Call the function using keyword arguments

def student_report(name, maths, science, english):
    total = maths + science + english
    average = total / 3
    return name, total, average

# call using keyword arguments
name,total,average = student_report(
    name="Dipam",
    maths=90,
    science=85,
    english=95
)

print("Name:", name)
print("Total Marks:", total)
print("Average:", average)

```

```python
# ‚úÖ Exercise 11 ‚Äî Grade Calculator Function
# Write a function get_grade(marks) that:

# Takes marks (0‚Äì100)

# Returns:
# "A" if marks ‚â• 90
# "B" if marks ‚â• 75
# "C" if marks ‚â• 60
# "D" if marks ‚â• 40
# "F" otherwise
# üëâ Sample:
# Input: 85
# Output: "B"

def get_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 40:
        return "D"
    else:
        return "F"

Totalmarks = int(input("Please enter marks: "))
grade = get_grade(Totalmarks)
print("Grade:", grade)
```

```python
# ‚úÖ Exercise 12 - Write a function to calculate factorial of a number
def factorial(num):
    result = 1
    while num > 0:
        result = result *num
        num -= 1
    return result

print(factorial(5))
```

```python
# ‚úÖ Exercise 13 ‚Äî Fibonacci Function
#üîπ Problem:
# Write a function fibonacci(n) that prints the first n numbers of the Fibonacci sequence.
#üîπ Reminder:
# Fibonacci starts like this:
# 0, 1, 1, 2, 3, 5, 8, 13, ...
# Each number = sum of previous two.

def fibonacci(n):
    a = 0
    b = 1
    count = 0
    
    while count < n:
        print(a)
        # calculate next number
        next_num = a + b
        a = b
        b = next_num
        count += 1

fibonacci(10)
```
