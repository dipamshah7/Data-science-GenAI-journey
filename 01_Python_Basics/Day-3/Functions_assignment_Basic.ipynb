```python
#Create a function that takes 2 numbers and returns both sum & product (Example: return 10+20 and 10*20)

def add_multiplication (num1,num2):

    return num1+num2, num1*num2

sum_val, product_val= add_multiplication(10,20)
print("sum of both number is",sum_val)
print("multiplication of both number is",product_val)
```

```python
#Create a function is_even(num) that returns True if even else False

def is_even(num):
    if num == 0:
        return "Zero is even"
    return num % 2 == 0

is_even(4)
```
