```python
# Basic Syntax

def greet():
    print("Hello!")

greet()
```

```python
# Function With Parameters
def greet(name):
    print("Hello", name)

greet("Dipam")
```

```python
# Return Statement
def add(a, b):
    return a + b

result = add(10, 20)

print(result)
```

```python
# Default Arguments
def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Dipam")
```

```python
# Keyword Arguments (order doesn't matter)

def student_info(name, age):
    print(name, age)

student_info(age=22, name="Raj")
```

```python
# Variable-Length Arguments
# *args → multiple positional arguments --> Works like tuple.

def total(*numbers):
    return sum(numbers)

print(total(10, 20, 30))

```

```python
# Variable-Length Arguments
# **kwargs → multiple keyword arguments --> Works like dictionary.

def info(**details):
    print(details)

info(name="Dipam", age=28)
```

```python
# Scope: Local vs Global Variables
def test():
    x = 10   # local
    print(x)

test()
```

```python
x = 100  # global

def test():
    print(x)

test()
```

```python
# Lambda (Anonymous Functions) --> Short 1-line functions.

square = lambda x: x * x

print(square(5))
```
