```python

# For loop
# Used when you know how many times you want to repeat.
fruits = ["apple", "banana", "mango"]

for item in fruits:
    print(item)
```

```python
#â–¶ range() explained
#range(start, stop, step)
#	â€¢ start â†’ where to begin
#	â€¢ stop â†’ where to stop (not included)
#   â€¢ step â†’ how much to increase

#ðŸ”¹ Loop 1 to 5
for i in range(1, 6):
    print(i)

#ðŸ”¹ Loop even numbers
for i in range(2, 11, 2):
    print(i)

#ðŸ”¹ Loop backwards
for i in range(10, 0, -1):
    print(i)
```

```python
num = 1
while num <= 5:
    print(num)
    num += 1
```

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```
