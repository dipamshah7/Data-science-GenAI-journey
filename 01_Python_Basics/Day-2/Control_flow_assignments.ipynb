```python
#Exercise 1

#Write a program:
#Input: temperature
#If temp > 30 â†’ "Hot"
#If 20â€“30 â†’ "Warm"
#If < 20 â†’ "Cold"

temp = int(input("Enter current temparature:"))

print("Current temparature:", temp)

if temp >= 30:
    print("Hot")
elif temp >= 20:
    print("Warm")
else:
    print ("Cold")
```

```python
#Exercise 2

#A shop gives discount:
#Above â‚¹1000 â†’ 20% off
#500â€“1000 â†’ 10% off
#Below 500 â†’ No discount
#Print final price.

order_value = int(input("Enter Bill amount: "))
print("Bill amount:", order_value)

if order_value >= 1000:
    discount_rate = 0.20
elif order_value >= 500:
    discount_rate = 0.10
else:
    discount_rate = 0

discount_amount = order_value * discount_rate
final_bill = order_value - discount_amount

print(f"Discount Applied: {discount_rate * 100}%")
print("Final Bill amount:", final_bill)
```

```python
#Exercise 3
#Check if a number is:
#Even
#Odd
#And also whether it is positive / negative

num = int(input("Enter a number: "))

if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero entered")

# Even / Odd check for all numbers
if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```
