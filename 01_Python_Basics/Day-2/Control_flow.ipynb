```python
# Basic IF
age = 20

if age >= 18:
    print("You are an adult.")

#✔ If condition is TRUE → code runs
#✔ If FALSE → code skipped
```

```python
# If Else statement
age = 16

if age >= 18:
    print("You can vote.")
else:
    print("You cannot vote.")
```

```python
# IF–ELIF–ELSE
# Use when you have multiple conditions.
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")

#✔ Python checks top to bottom
#✔ First TRUE condition executes
```

```python
# NESTED IF (IF inside IF)

age = 18
has_id = True

if age >= 18:
    if has_id:
        print("Entry allowed")
    else:
        print("ID required")
else:
    print("Entry denied")
```

```python
#LOGICAL OPERATORS

#You use these inside conditions:

# Operator  <-->	Meaning
# and       <-->	Both conditions must be true
# or        <-->	At least one must be true
# not       <-->	Reverse the condition

age = 22
citizen = True

if age >= 18 and citizen:
    print("Eligible to vote")
```
