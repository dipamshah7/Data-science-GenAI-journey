```python
#Exercise 3

#Take a number from user and print whether each number from 1 to that number is Even or Odd.

num = int(input("Enter a number:"))

for i in range (1,num+1):
    if i%2 ==0:
        print(i,"Even")
    else:
        print(i,"Odd")
```

```python
#EXERCISE 6 — Count Up to N

#Ask the user for a number N and print numbers from 1 to N using a while loop.

num = int(input("Enter numnber:"))
i = 1
while i <= num :
    print(i)
    i+=1
```

```python
#EXERCISE 7 — Sum Until User Enters 0

#Keep taking numbers from user and add to total.
#Stop only when user enters 0.

total = 0

while True:
    num = int(input("Enter a number: "))
    
    if num == 0:
        break
    
    total += num

print("Total:", total)
```

```python
#EXERCISE 9 — Countdown Timer

#Ask for a start number.
#Count down to 0 using while.

start = int(input("Enter start time:"))

while start >=0 :
    print(start)
    start -= 1
```

```python
#EXERCISE 10 — Multiplication Table Using While

#Take a number from user (say 9).
#Print table of 9 using while loop.

num = int(input("Enter a number:"))

mult=1

while mult <=10:
    total = num*mult
    print(num,"x",mult,"=",total)
    mult += 1
```

```python
#EXERCISE 11 — Largest Number Finder

#User keeps entering numbers.
#Stop only when user types "done".

#At the end, print the largest number entered.

largest = None

while True:
    num = input("Enter a number: ")

    if num.lower() == "done":
        print("Largest number:", largest)
        break

    num = int(num)

    if largest is None or num > largest:
        largest = num
```

```python
#Exercise 12 — Sum of Even & Odd Numbers Task
#Keep asking the user to enter numbers.
#Stop only when user types "stop".
#Add all even numbers into one sum.
#Add all odd numbers into another sum.
#At the end, print:
#Total sum of even numbers
#Total sum of odd numbers

odd=[]
even=[]

while True:
    num = input("Enter a number (or type 'stop'): ")

    print("Entered number:",num)

    if num.lower() == "stop":
        print("Total sum of even numbers:", sum(even))
        print("Total sum of odd numbers:", sum(odd))
        break

    num =int(num)

    if num%2 == 0:
        even.append(num)
    else:
        odd.append(num)
```
