```python
#Dictionaries
car = {
    "brand": "Mercedes",
    "model": "C-Class",
    "year": 2023,
    "owner": "Dipam"
}

print(car)
print(type(car))

print(car["brand"])        # direct access
print(car.get("year"))     # safe access

car["year"] = 2025   # update
car["fuel"] = "Petrol"  # add new key
print(car)

car.pop("model")
print(car)

for key, value in car.items():
    print(key, ":", value)
```

```python
#Mini Dictionary Exercise
student = {
    "name": "Dipam",
    "age": 28,
    "branch": "Mechanical",
    "company": "Mercedes",
    "skills": ["Python", "Design", "CAE"]
}

print(student.get("name"))
print(len(student.get("skills")))

student.get("skills").append("Data Science")
print(student)

student["company"] = "Data Science Aspirant"
print(student)

print(student.get("skills").pop(1))
print(student)

student.update({"age": 29, "branch": "AI"})
```
