```python
#list
numbers = [10, 20, 30, 40, 50]

print(numbers)
print(numbers[0])
print(numbers[-1])

numbers.append(60)
print(numbers)

numbers.remove(30)
print(numbers)

print("Length:", len(numbers))
```

```python
#list Excersise
marks = [78, 92, 64, 88, 73]

#Print highest mark
highest = max(marks)
print("Highest Marks:", highest)

#Print lowest mark
lowest = min(marks)
print("Lowest Marks:", lowest)

#Print average
Total_marks = sum(marks)
no_of_students = len(marks)
average = Total_marks/no_of_students
print("Average Marks:", average)

#Add a new mark 95
marks.append(95)
print(marks)

#Remove the lowest mark
marks.remove(lowest)
print(marks)
```
