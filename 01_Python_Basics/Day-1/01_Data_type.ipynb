```python
#Data Type
name = "Dipam"
age = 28
height = 5.9
is_engineer = True

print(name, age, height, is_engineer)

print(type(name))       # string
print(type(age))        # integer
print(type(height))     # float
print(type(is_engineer))  # boolean
```

```python
#Operations
a = 12
b = 4

print("Add:", a + b)
print("Subtract:", a - b)
print("Multiply:", a * b)
print("Divide:", a / b)
print("Integer Divide:", a // b)
print("Power:", a ** b)
```

```python
#Strings
message = "Data Science Learning"

print(message.upper())
print(message.lower())
print(message.split())
print(message[0:4])   # slicing
print(message[::-1]) # reverse string
```
